```python
import pandas as pd
```

```python
df = pd.read_excel('../zillow/input2/SeriesReport-20200524095221_9a2f67.xlsx', skiprows=19).iloc[:,:-1]
df
```

```python
df = pd.melt(df,
            id_vars = 'Year',   # all columns before dates
            #value_vars= dfx_ord.columns[QTr_1:], # columns with dates
            var_name='quarter',
            value_name='wages').sort_values(['Year', 'quarter']).dropna()
```

```python
#Add date column
dic = {'Qtr1':'-01-01',
           'Qtr2':'-04-01',
           'Qtr3':'-07-01',
           'Qtr4':'-10-01'}
dic
```

```python
df.quarter = df.quarter.map(dic)
```

```python
# concatenate dates - with same logic of dates imported from FRED (start of period)
df['DATE'] = df.Year.astype(str).str.cat(df.quarter)

# drop columns
df.drop(['Year', 'quarter'],1, inplace= True)
```

```python
# transform wages into annual income
df.wages = df.wages * 52
```
